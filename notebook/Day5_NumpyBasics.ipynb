## Numpy basics


```python
import numpy as np
a=np.array([[1,2,3],[2,3,4]])
b=np.array([[2,3,1],[9,8,7]])
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a**b)
print(a/b)
print(a//b)
print(a%b)
```

```python
print(a.max())
print(a.mean())
print(np.median(b))
```

```python
b=b.transpose()
np.dot(a,b)
```
