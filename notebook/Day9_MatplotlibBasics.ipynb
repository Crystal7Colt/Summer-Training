```python
import matplotlib.pyplot as plt
x=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
y=[23,31,42,13,14,12]
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Report")
plt.plot(x,y,color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

x1=[2,6,12]
y1=[2,2,6]
x2=[2,6,12]
y2=[7,7,11]
x3=[4,10]
y3=[10,14]
x4=[2,2,4]
y4=[2,7,10]
x5=[6,6,4]
y5=[2,7,10]
x6=[12,12,10]
y6=[6,11,14]
x7=[3,3,4.5,4.5]
y7=[2,5,5,2]
Circle=plt.Circle((3.75,8),0.5)
fig,ax=plt.subplots()
ax.add_patch(Circle)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x4,y4)
plt.plot(x5,y5)
plt.plot(x6,y6)
plt.plot(x7,y7)
plt.title("House")
plt.show()
```
