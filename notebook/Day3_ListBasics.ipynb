```python
import pandas as pd
df=pd.read_csv("WC_players.csv")
data=df['Player']
country=df['Country']
data=data.tolist()
country=country.tolist()
a=True
unique=[]
for i in country:
    if i not in unique:
        unique.append(i)
while(a):
    
    print("What do you want to check")
    print("Press 1 to check participating countries")
    print("Press 2 to Check players from a specific country")
    print("Press 3 to exit")
    Press = int(input("Enter your choice"))
    if Press==3:
        a=False
    elif Press==1:
        print(unique)
    elif Press == 2:
        print("Which country")
        for i in range(len(unique)):
            print(i , end=" ")
            print(unique[i])
        Selection=int(input("choose the country number"))
        count=-1
        for i in country:
            count=count+1
            if i==unique[Selection]:
                print(data[count])

```
