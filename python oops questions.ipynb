Q1. Student Information System
Create a Student class with attributes: Name, Roll Number, Marks (3 subjects).
Methods: get_data(), display_data(), calculate_percentage().

```python
class Student:
    def __init__(self,name,roll_number):
        self.name = name
        self.roll_number = roll_number
    def get_data(self,physics_marks,chemistry_marks,biology_marks):
        self.physics_marks= physics_marks
        self.chemistry_marks= chemistry_marks
        self.biology_marks = biology_marks
    def display_data(self):
        print(f"The student's name is {self.name} and has rollno {self.roll_number}")
        print(f"He got {self.physics_marks} in physics, {self.chemistry_marks} in chemistry, {self.biology_marks} in biology")
    def calculate_percentage(self):
        self.perct = (self.physics_marks + self.chemistry_marks + self.biology_marks)/3
        print(f"{self.name} has a percentage of {self.perct}")

s1 = Student("Kaustuv",2)
s1.get_data(70,90,80)
s1.display_data()
s1.calculate_percentage()
```

Q3. Employee Salary Calculator
Create an Employee class with Employee ID, Name, Basic Salary.
Methods to calculate HRA (20%), DA (10%), Gross Salary, and display details.

```python
class Employee:
    def __init__(self,ID,Name,Salary):
        self.id=ID
        self.name = Name
        self.salary = Salary
        self.hra = 0.0
        self.da = 0.0
        self.gross = 0.0
    def HRA(self):
        self.hra= 0.2*self.salary
        print(f"The HRA received is {self.hra}")
    def DA(self):
        self.da=0.1*self.salary
        print(f"The DA received is {self.da}")
    def Gross(self):
        self.gross=self.hra+self.da+self.salary
        print(f"The Gross salary received is {self.gross}")
    def display_detail(self):
        print(f"Employee name is {self.name} \nEmployee ID is {self.id} \nEmployee basic salary is {self.salary}")
        print(f"The HRA received is {self.hra}\nThe DA received is {self.da}\nThe Gross salary received is {self.gross}")

Emp1 = Employee(101,"Kaustuv",1000)
Emp1.DA()
Emp1.HRA()
Emp1.Gross()
Emp1.display_detail()
```

Q4. Rectangle Operations
Create a Rectangle class with Length and Breadth.
Methods: area(), perimeter(), check_square().

```python
class Rectangle:
    def __init__(self,length,breadth):
        self.length = length
        self.breadth = breadth
    def Area(self):
        self.area = self.length * self.breadth
        print(f"The area of the Rectangle is {self.area}")
    def Perimeter(self):
        self.perimeter = 2*(self.length+self.breadth)
        print(f"The Perimeter of the rectangle is {self.perimeter}")
    def Check_square(self):
        if self.length==self.breadth:
            print("The Rectangle is also a Square")
        else:
            print("The Rectangle is not a square")
rect1= Rectangle(12,15)
rect2 = Rectangle(12,12)
rect1.Area()
rect1.Perimeter()
rect1.Check_square()
rect2.Area()
rect2.Perimeter()
rect2.Check_square()
```

Q8. Payment Gateway
Create Payment parent class.
Child classes: CreditCard, UPI, NetBanking.
Override pay(amount) method.

```python
class PaymentGateway:
    def __init__(self,wallet):
        self.balance = wallet
    def pay(self,amount):
        self.amount= amount
        print(f"Paid {self.amount}")
class CreditCard(PaymentGateway):
    def __init__(self, wallet):
        super().__init__(wallet)
    def pay(self, amount):
        self.amount = amount
        print(f"Paid {self.amount+(self.amount*0.02)} with 2% handling fees")
class UPI(PaymentGateway):
    def __init__(self, wallet):
        super().__init__(wallet)
    def pay(self, amount):
        self.amount = amount
        if self.amount<= self.balance:
            print(f"Paid {self.amount}")
        else:
            print("insufficient wallet balance")
class NetBanking(PaymentGateway):
    def __init__(self, wallet):
        super().__init__(wallet)
    def pay(self, amount):
        self.amount= amount
        if self.amount<= self.balance:
            print(f"Paid {self.amount}")
        else:
            print("Transaction failed")

cc= CreditCard(1000)
cc.pay(500)
upi=UPI(1500)
upi.pay(700)
nb = NetBanking(5000)
nb.pay(2000)
```

Q10. ATM Machine (Encapsulation)
Create ATM class with private PIN and Balance.
Methods: verify_pin(), deposit(), withdraw(), change_pin(), display_balance().
PIN verification required before every transaction.

```python
class AtmCard:
    def __init__(self, cardnumber , pin, balance):
        self.cardnumber = cardnumber
        self.__pin = pin
        self.balance = balance
    def ShowBalance(self):
        print(self.balance)
    def Withdraw(self,amount):
        if amount<=self.balance:
            self.balance-=amount
            print(f"Amount withdrawn is {amount}, the new balance is {self.balance}")
        else:
            print("Not Enough balance in Account")
    def Deposit(self,amount):
        self.balance+=amount
        print(f"Amount Deposited is {amount}, the new balance is {self.balance}")
    def Changepin(self,Newpin):
        self.pin = Newpin
        print(f"The New pin is set to{Newpin}")
    def Verify_pin(self,enteredpin):
        if enteredpin==self.__pin:
            return 1
        else:
            return 0


Card1 = AtmCard(1001001,1947,1000)
Card2 = AtmCard(1001002,1936,1000)
Card3 = AtmCard(1001003,1911,1000)
Atmdict={Card1.cardnumber:Card1,Card2.cardnumber:Card2,Card3.cardnumber:Card3}
Card = int(input("Enter your Card Number "))
if Card in Atmdict:
    Pin=int(input("Enter the pin "))
    active=Atmdict[Card]
    if active.Verify_pin(Pin):
        while(True):
            print("What do you want to perform \nPress 1 to check balance \
            \nPress 2 to Deposit \nPress 3 to Withdraw \nPress 4 to change pin \nPress 5 to exit ")
            Press = int(input("Enter your choice "))
            match Press:
                case 1:
                    active.ShowBalance()
                case 2:
                    Pin=int(input("Enter the pin "))
                    if active.Verify_pin(Pin):
                        dep=int(input("Enter the Amount you want to Deposit "))
                        active.Deposit(amount=dep)
                    else: print("Transcation Failed")
                case 3:
                    Pin=int(input("Enter the pin "))
                    if active.Verify_pin(Pin):
                        wit=int(input("Enter the Amount you want to Withdraw "))
                        active.Withdraw(amount=wit)
                    else: print("Transcation Failed")
                case 4:
                    Np=int(input("Enter the New pin "))
                    active.Changepin(Np)
                case 5:
                    break
    else: print("incorrect pin")
else:
    print("invalid card")
```

Q11. Employee Payroll System (Abstraction)
Using abc module, create abstract Employee class with calculate_salary() and display().
Child classes: FullTimeEmployee and PartTimeEmployee.

```python
from abc import ABC , abstractmethod
class Employee(ABC):
    @abstractmethod
    def Calculate_salary():
        pass
    @abstractmethod
    def display():
        pass

class FullTimeEmployee(Employee):
    def __init__(self,Name,Salary):
        self.name = Name
        self.salary = Salary
    def Calculate_salary(self):
        self.gross=self.salary+(self.salary*0.3)
        print(f"The Gross salary received is {self.gross}")
    def display(self):
        print(f"Employee name is {self.name} \nEmployee basic salary is {self.salary}")
        print(f"The HRA received is {self.salary*0.2}\nThe DA received is {self.salary*0.1}\nThe Gross salary received is {self.gross}")

class PartTimeEmployee(Employee):
    def __init__(self,Name,Salary):
        self.name = Name
        self.salary = Salary
    def Calculate_salary(self):
        print(f"The Gross salary received is {self.salary}")
    def display(self):
        print(f"Employee name is {self.name} \nEmployee salary is {self.salary}")

Emp1 = FullTimeEmployee("Kaustuv",1000)
Emp1.Calculate_salary()
Emp1.display()
Emp2 = PartTimeEmployee("Laxman",1000)
Emp2.Calculate_salary()
Emp2.display()
```

Q12. Online Shopping System
Create abstract Product class with private Product ID and Price.
Child classes: Electronics (10%), Clothing (20%), Grocery (5%).
Implement inheritance, encapsulation, abstraction and polymorphism

```python
from abc import ABC ,abstractmethod
class Product(ABC):
    def __init__(self,ID,Price):
        self.__id=ID
        self.price=Price
    @abstractmethod
    def discount(self):
        pass
class Electronics(Product):
    def __init__(self, ID, Price):
        super().__init__(ID, Price)
    def  discount(self):
        self.price = self.price-(self.price*0.1)
        print(f"Final price is {self.price}")
class CLothing(Product):
    def __init__(self, ID, Price):
        super().__init__(ID, Price)
    def  discount(self):
        self.price = self.price-(self.price*0.2)
        print(f"Final price is {self.price}")
class Grocery(Product):
    def __init__(self, ID, Price):
        super().__init__(ID, Price)
    def  discount(self):
        self.price = self.price-(self.price*0.05)
        print(f"Final price is {self.price}")

Tv = Electronics(101,15000)
Shirt = CLothing(201,1500)
Bread = Grocery(301,55)

Tv.discount()
Shirt.discount()
Bread.discount()
```
